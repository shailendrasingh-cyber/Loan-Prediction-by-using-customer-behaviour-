```python
# import teh library 
import seaborn as sns 
```

```python
# importing the library 
import matplotlib.pyplot as plt 
```

```python
# EDA on married/single columns 
```

```python
df['Married/Single'].value_counts()
```

```python
df['Married/Single'].value_counts().plot(kind='bar')
```

There are more single people in our data as compared to married people 

```python
plt.Figure(figsize=(15,6))
ax = sns.countplot(data = df , x=df["Married/Single"], hue= "Risk_Flag" )
plt.title('Distribution of the target variable according to Martial Status ')
for p in ax.containers:
    ax.bar_label(p)
```

```python
# EDA on House_Ownership
```

```python
df['House_Ownership'].value_counts()
```

```python
df['House_Ownership'].value_counts().plot(kind='bar')
```

The more people are living in rented house as compared to owned and nonrent_nonown

```python
plt.Figure(figsize=(15,6))
ax = sns.countplot(data=df , x = df['House_Ownership'] , hue='Risk_Flag')
plt.title('Distribution of the target variable according to House Ownership ')
for p in ax.containers:
    ax.bar_label(p)
```
